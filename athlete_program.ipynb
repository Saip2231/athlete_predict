```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

date_rng = pd.date_range(start="2020-01-01", end="2022-12-31", freq="M")
performance_scores = np.random.randint(50, 100, size=len(date_rng))
df = pd.DataFrame({"Date": date_rng, "Performance_Score": performance_scores})
df.set_index("Date", inplace=True)

plt.figure(figsize=(10, 6))
plt.plot(df.index, df["Performance_Score"], label="Performance Score")
plt.title("Time-Series Data")
plt.xlabel("Date")
plt.ylabel("Performance Score")
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error

train_size = int(len(df) * 0.8)
train, test = df[:train_size], df[train_size:]

model = ARIMA(train, order=(5, 1, 0))
model_fit = model.fit()

forecast = model_fit.forecast(steps=len(test))
test["Predicted"] = forecast

plt.figure(figsize=(10, 6))
plt.plot(train.index, train["Performance_Score"], label="Train")
plt.plot(test.index, test["Performance_Score"], label="Test", color="blue")
plt.plot(test.index, test["Predicted"], label="Predicted", color="red")
plt.title("ARIMA Model")
plt.xlabel("Date")
plt.ylabel("Performance Score")
plt.legend()
plt.show()

mse = mean_squared_error(test["Performance_Score"], test["Predicted"])
print(f"Mean Squared Error (ARIMA): {mse}")
```

```python
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import TimeSeriesSplit
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.preprocessing import MinMaxScaler

data = {
    "Date": pd.date_range(start="2023-01-01", periods=100, freq="D"),
    "Performance": np.sin(np.linspace(0, 50, 100)) + np.random.normal(0, 0.1, 100)
}
df = pd.DataFrame(data)

df["Date"] = pd.to_datetime(df["Date"])
df.set_index("Date", inplace=True)
performance_data = df["Performance"].values.reshape(-1, 1)

scaler = MinMaxScaler()
performance_data_scaled = scaler.fit_transform(performance_data)

def create_sequences(data, sequence_length):
    X, y = [], []
    for i in range(len(data) - sequence_length):
        X.append(data[i:i+sequence_length])
        y.append(data[i+sequence_length])
    return np.array(X), np.array(y)

sequence_length = 10
X, y = create_sequences(performance_data_scaled, sequence_length)

train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

def build_lstm_model(input_shape):
    model = Sequential([
        LSTM(50, activation="relu", input_shape=input_shape),
        Dense(1)
    ])
    model.compile(optimizer="adam", loss="mse")
    return model

def evaluate_metrics(y_true, y_pred):
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true, y_pred)
    
    print(f"MSE: {mse}")
    print(f"RMSE: {rmse}")
    print(f"R-squared: {r2}")
    

def cross_validate_lstm(X, y, n_splits=5):
    tscv = TimeSeriesSplit(n_splits=n_splits)
    mse_scores = []

    for train_index, test_index in tscv.split(X):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]
        
       
        model = build_lstm_model((X_train.shape[1], 1)) 
        model.fit(X_train, y_train, epochs=20, batch_size=16, verbose=0)
        
     
        y_pred = model.predict(X_test)
        mse_scores.append(mean_squared_error(y_test, y_pred))
        print(f"Cross-Validation MSE (mean): {np.mean(mse_scores)}")
    
evaluate_metrics(y_test, model.predict(X_test))

cross_validate_lstm(X, y)

def prediction_confidence_intervals(model, X, y, alpha=0.05):
   
    y_pred = model.predict(X)
    
    
    residuals = y - y_pred
    residual_std = np.std(residuals)
    
    
    z_score = 1.96  # 95% confidence
    ci_lower = y_pred - z_score * residual_std
    ci_upper = y_pred + z_score * residual_std
    
    return ci_lower, ci_upper


ci_lower, ci_upper = prediction_confidence_intervals(model, X_test, y_test)


plt.figure(figsize=(10, 6))
plt.plot(df.index[-len(y_test):], y_test, label="Actual")
plt.plot(df.index[-len(y_test):], model.predict(X_test), label="Predicted")
plt.fill_between(df.index[-len(y_test):], ci_lower.flatten(), ci_upper.flatten(), color='gray', alpha=0.2, label="95% Confidence Interval")
plt.title("LSTM Predictions with Confidence Interval")
plt.xlabel("Date")
plt.ylabel("Performance")
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

data = {
    "Date": pd.date_range(start="2023-01-01", periods=100, freq="D"),
    "Performance": np.sin(np.linspace(0, 50, 100)) + np.random.normal(0, 0.1, 100)
}
df = pd.DataFrame(data)

df["Date"] = pd.to_datetime(df["Date"])
df.set_index("Date", inplace=True)
performance_data = df["Performance"].values.reshape(-1, 1)

scaler = MinMaxScaler()
performance_data_scaled = scaler.fit_transform(performance_data)

def create_sequences(data, sequence_length):
    X, y = [], []
    for i in range(len(data) - sequence_length):
        X.append(data[i:i+sequence_length])
        y.append(data[i+sequence_length])
    return np.array(X), np.array(y)

sequence_length = 10
X, y = create_sequences(performance_data_scaled, sequence_length)

train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

def build_lstm_model(input_shape):
    model = Sequential([
        LSTM(50, activation="relu", input_shape=input_shape),
        Dense(1)
    ])
    model.compile(optimizer="adam", loss="mse")
    return model

model = build_lstm_model((X_train.shape[1], 1))
history = model.fit(X_train, y_train, epochs=20, batch_size=16, validation_data=(X_test, y_test))

y_pred = model.predict(X_test)
y_pred_rescaled = scaler.inverse_transform(y_pred)
y_test_rescaled = scaler.inverse_transform(y_test)

def evaluate_metrics(y_true, y_pred):
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true, y_pred)
    
    print(f"MSE: {mse}")
    print(f"RMSE: {rmse}")
    print(f"R-squared: {r2}")
    
evaluate_metrics(y_test_rescaled, y_pred_rescaled)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(df.index[-len(y_test_rescaled):], y_test_rescaled, label="Actual")
plt.plot(df.index[-len(y_test_rescaled):], y_pred_rescaled, label="Predicted")
plt.title("LSTM Predictions vs Actual")
plt.xlabel("Date")
plt.ylabel("Performance")
plt.legend()
plt.show()
```
